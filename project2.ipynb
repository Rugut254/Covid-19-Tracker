```python
import pandas as pd
df = pd.read_csv('owid-covid-data.csv')
print(df.describe())
```

```python
df.fillna(0, inplace=True)
```

```python
countrywise_stats = df.groupby("location")["total_cases"].sum().reset_index()
print(countrywise_stats.sort_values(by="total_cases", ascending=False))
```

```python
daily_trends = df.groupby("location")["total_cases"].sum()
print(daily_trends.tail(255))  
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=df["location"], y=df["total_cases"])
plt.title("Global COVID-19 Cases Over Time")
plt.xlabel('location')
plt.ylabel("total_cases")
plt.show()
```

```python
continent_cases = df.groupby("continent")["total_cases"].sum().reset_index()

print(continent_cases)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x="continent", y="total_cases", data=continent_cases, palette="coolwarm")

plt.title("Total COVID-19 Cases by Continent")
plt.xlabel("Continent")
plt.ylabel("total_cases")
plt.xticks(rotation=45)

plt.show()
```
